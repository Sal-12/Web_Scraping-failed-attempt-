```python
url = 'https://en.wikipedia.org/wiki/ProgrammingLanguage'
page = requests.get(url)
content = page.content
#print(content)
soup = BeautifulSoup(content, 'html.parser')
print(soup)
#print(soup.prettify)
#tree traverse

```
